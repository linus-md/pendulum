```python
import numpy as np
import matplotlib.pyplot as plt

# Set the range for x and y coordinates
x, y = np.meshgrid(np.linspace(-100, 100, 20), np.linspace(-100, 100, 20))

# Define the function for the vector field: F(x, y) = (-y, x)
u = x
v = y - 1

# Plotting the vector field
plt.figure(figsize=(8, 8))
plt.quiver(x, y, u, v)

plt.xlim(-100, 100)
plt.ylim(-100, 100)
#plt.xlabel('X axis')
#plt.ylabel('Y axis')
#plt.title('Vector Field of F(x, y) = (x, y-1)')
plt.grid(False)
plt.show()
```
